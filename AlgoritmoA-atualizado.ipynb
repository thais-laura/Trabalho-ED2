```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit


x = [100, 500, 1000, 5000, 10000]
y1 = [0.199, 10.47, 37.50, 1177.1, 4997.5]
y2 = [0.497, 10.47, 60.94, 1379.2, 5752.4]
y3 = [0, 9.375, 57.25, 1440.5, 5479.5]
y4 = [0.595, 13.76, 47.07, 1242.4, 5286.8]

media = [(100, 0.32275), (500, 11.01875), (1000, 51.19), (5000, 1309.8), (10000, 5379.05)]
x_med, y_med = zip(*media)

# Regress√£o linear a partir dos valores m√©dios
def regression_function_lin(x, a, b):
    return a * x + b
def regression_function_nlog(x, a, b):
    return a * x * np.log(x) + b
def regression_function_quad(x, a, b, c):
    return a * x**2 + b * x + c

params, covariance = curve_fit(regression_function_lin, x_med, y_med)
a, b = params
x_lin = np.linspace(min(x_med), max(x_med), 100)
y_lin = regression_function_lin(x_lin, a, b)
print("Equacao linear: y = ", a,"*x +", b)

params, covariance = curve_fit(regression_function_nlog, x_med, y_med)
a, b = params
x_nlog = np.linspace(min(x_med), max(x_med), 100)
y_nlog = regression_function_nlog(x_nlog, a, b)
print("Equacao n*log(n): y = ", a, "*x*log(x) +", b)


params, covariance = curve_fit(regression_function_quad, x_med, y_med)
a, b, c = params
x_quad = np.linspace(min(x_med), max(x_med), 100)
y_quad = regression_function_quad(x_quad, a, b, c)
print("Equacao quadr√°tica: y = ", a, "*x^2 +", b, "x +", c)


plt.scatter(x, y1)
plt.scatter(x, y2)
plt.scatter(x, y3)
plt.scatter(x, y4)
plt.scatter(x_med, y_med)
plt.plot(x_lin, y_lin, color='purple')
plt.plot(x_nlog, y_nlog, color='gray')
plt.plot(x_quad, y_quad, color='yellow')

plt.xlabel('Tamanho da entrada')
plt.ylabel('Dura√ß√£o [ms]')
plt.title('Algoritmo A')
plt.legend(['Ord cresc', 'Ord decresc', 'Aleat', 'Aleat repet','Dados m√©dios', 'Regress√£o Linear', 'Regress√£o n*log(n)', 'Regressao quadratica'])
plt.grid(True)
plt.show()


fig, axs = plt.subplots(3, 1, figsize=(10, 15))

axs[0].scatter(x, y1, label='y1')
axs[0].scatter(x, y2, label='y2')
axs[0].scatter(x, y3, label='y3')
axs[0].scatter(x, y4, label='y4')
axs[0].scatter(x_med, y_med, label='Dados m√©dios')
axs[0].plot(x_lin, y_lin, color='purple', label='Regress√£o linear')
axs[0].set_xlabel('Tamanho da entrada')
axs[0].set_ylabel('Dura√ß√£o [ms]')
axs[0].set_title('Algoritmo A')
axs[0].legend()
axs[0].grid(True)

axs[1].scatter(x, y1, label='y1')
axs[1].scatter(x, y2, label='y2')
axs[1].scatter(x, y3, label='y3')
axs[1].scatter(x, y4, label='y4')
axs[1].scatter(x_med, y_med, label='Dados m√©dios')
axs[1].plot(x_nlog, y_nlog, color='gray', label='Regress√£o n*log(n)')
axs[1].set_xlabel('Tamanho da entrada')
axs[1].set_ylabel('Dura√ß√£o [ms]')
axs[1].set_title('Algoritmo A')
axs[1].legend()
axs[1].grid(True)

axs[2].scatter(x, y1, label='y1')
axs[2].scatter(x, y2, label='y2')
axs[2].scatter(x, y3, label='y3')
axs[2].scatter(x, y4, label='y4')
axs[2].scatter(x_med, y_med, label='Dados m√©dios')
axs[2].plot(x_quad, y_quad, color='gold', label='Regress√£o quadr√°tica')
axs[2].set_xlabel('Tamanho da entrada')
axs[2].set_ylabel('Dura√ß√£o [ms]')
axs[2].set_title('Algoritmo A')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
